## Making a Plot

With the data in hand, let's go ahead and make a quick plot. First, we will read in the data:

````python
n_rows = 231
n_cols = 381
variables = ['THETA','SALT']
data = np.fromfile(os.path.join(run_dir,'diags','TS_surf_daily_mean','TS_surf_daily_mean.0000000156.data'),'>f4')
data = np.reshape(data,(len(variables),n_rows,n_cols))
````

Next, let's make a plot:

````python
plt.figure()
C = plt.pcolormesh(data[0,:,:],cmap='turbo', vmin=-5, vmax=15)
plt.colorbar(C)
plt.show()
plt.savefig("my_plot.png", dpi=300, bbox_inches='tight')
````
