```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nx, ny, nz = 77, 127, 50
depth_levels = np.linspace(0, -4000, nz)
```

```python
# Create dummy initial conditions
THETA = np.zeros((nz, ny, nx))  # Temperature
SALT = np.ones((nz, ny, nx)) * 35  # Salinity at 35 PSU
UVEL = np.zeros((nz, ny, nx))  # Zonal Velocity (u)
VVEL = np.zeros((nz, ny, nx))  # Meridional Velocity (v)
SSH = np.zeros((ny, nx))  # Sea Surface Height
```

```python
# Add variations 
for k in range(nz):
    THETA[k, :, :] = -2 + (depth_levels[k] / 1000)  

THETA.astype('>f4').tofile("THETA_init.bin")
SALT.astype('>f4').tofile("SALT_init.bin")
UVEL.astype('>f4').tofile("UVEL_init.bin")
VVEL.astype('>f4').tofile("VVEL_init.bin")
SSH.astype('>f4').tofile("SSH_init.bin")
```

```python
# Initial Conditions Plots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

c1 = axs[0].imshow(THETA[0, :, :], cmap="coolwarm", origin="lower")
axs[0].set_title("Initial Temperature at Surface")
fig.colorbar(c1, ax=axs[0])

c2 = axs[1].imshow(SALT[0, :, :], cmap="viridis", origin="lower")
axs[1].set_title("Initial Salinity at Surface")
fig.colorbar(c2, ax=axs[1])

c3 = axs[2].imshow(SSH, cmap="Blues", origin="lower")
axs[2].set_title("Initial Sea Surface Height")
fig.colorbar(c3, ax=axs[2])

plt.tight_layout()
plt.show()
```
